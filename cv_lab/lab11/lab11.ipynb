```
import numpy as np
import cv2 as cv
from skimage import img_as_float
from PIL import Image
import time

def show(img):
    if (len(img.shape) == 2):
        display(Image.fromarray(img));    
    else:
        display(Image.fromarray(img[:,:,::-1]));
filename = 'chess.png'
img = cv.imread(filename)
img=cv.resize(img, (500,500), interpolation = cv.INTER_AREA)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
gray = np.float32(gray)
time1=time.time()
dst = cv.cornerHarris(gray,2,3,0.04)
dst = cv.dilate(dst,None)
img[dst>0.05*dst.max()]=[0,0,255] 
time2=time.time()-time1
print("time="+str(time2))
#image = cv.putText(img, 'OpenCV', (15,15),2 , cv.FONT_HERSHEY_SIMPLEX, (255, 0, 0), 2, cv.LINE_AA)
cv.imshow( 'output',img)
if cv.waitKey(0) & 0xff == 27:
    cv.imwrite('chessout.jpg', img)
    cv.destroyAllWindows() 
show(img)
```
