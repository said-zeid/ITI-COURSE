```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
def SHOW(img):
    display(Image.fromarray(img[:,:,::-1]));
img = cv2.imread('perspective02.jpg')
rows,cols = img.shape[:2]
pts1 = np.float32(
    [
        [1,  719], [1279, 719],
        [580, 340], [772, 340]
    ]
)

pts2 = np.float32(
    [
        [580, 720], [772,720],
        [580, 0], [772,0]
    ]
)

M = cv2.getPerspectiveTransform(pts1 ,pts2)
dst = cv2.warpPerspective(img, M, (cols,rows))

SHOW(dst)
#plt.imshow(img[:,:,::-1]);
img.shape

```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def SHOW(img):
    display(Image.fromarray(img[:,:,::-1]));
img = cv2.imread('perspective01.png')
rows,cols = img.shape[:2]
pts1 = np.float32(
    [
        [240,  200], [600, 200],
        [0, 240], [960, 300]
    ]
)

pts2 = np.float32(
    [
        [240, 0], [600,0],
        [249, 540], [600,540]
    ]
)

M = cv2.getPerspectiveTransform(pts1 ,pts2)
dst = cv2.warpPerspective(img, M, (cols,rows))

SHOW(dst)
#plt.imshow(img[:,:,::-1]);
img.shape
```
