```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('carkyo.jpg')
roi = img[300:450, 300:450]


logo = cv2.imread('tekomoro.png')
logo = cv2.resize(logo, (150, 34))
roi = img[724:758,1205:1355];
copy=img.copy()
out = img.copy()
gray = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
mask_inv = cv2.bitwise_not(mask)

img_bg = cv2.bitwise_and(roi, roi, mask=mask)
img_fg = cv2.bitwise_and(logo, logo, mask=mask_inv)
dst = cv2.add(img_bg, img_fg)
out[724:758,1205:1355]=dst
plt.imshow(out[:,:,::-1]);





#img_fg = cv2.bitwise_and(logo, logo, mask=mask_inv)
#out[724:758,1205:1355] = cv2.addWeighted(copy[724:758,1205:1355], 1, logo, 1, 0)


```
